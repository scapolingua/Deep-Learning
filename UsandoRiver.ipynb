```python
# Codice Nuovo aggiustato
from river import forest, metrics, preprocessing
import numpy as np  # Import numpy for handling nan
import pandas as pd

# Carica i dati
consumi_elettrici_2023 = pd.read_csv('consumi_elettrodomestici_2023.csv', sep=';')
consumi_elettrici_2023['Data'] = pd.to_datetime(consumi_elettrici_2023['Data'], format='%d-%m-%Y')
consumi_elettrici_2023['Ora'] = pd.to_datetime(consumi_elettrici_2023['Ora'], format='%H:%M')
consumi_elettrici_2023['Anno'] = consumi_elettrici_2023['Data'].dt.year
consumi_elettrici_2023['Mese'] = consumi_elettrici_2023['Data'].dt.month
consumi_elettrici_2023['Giorno'] = consumi_elettrici_2023['Data'].dt.day
consumi_elettrici_2023['Ore'] = consumi_elettrici_2023['Ora'].dt.hour
consumi_elettrici_2023['Minuti'] = consumi_elettrici_2023['Ora'].dt.minute
consumi_elettrici_2023 = consumi_elettrici_2023.drop(['Data', 'Ora'], axis=1)

# Use ARFRegressor for continuous target
model = forest.ARFRegressor(seed=8)

metric = metrics.MAE()  # Mean Absolute Error

# Training and Prediction Loop
predictions = []
for i, row in consumi_elettrici_2023.iterrows():
    x = row.drop('TotaleConsumo', errors='ignore').to_dict()

    # Handle potential missing values in x
    for key, value in x.items():
        if pd.isnull(value) or np.isnan(value):  # Check for both pandas and numpy nan
            x[key] = 0  # Replace nan with 0 as an example

    y = row['TotaleConsumo']  # Assuming this is your target

    # Basic Logging for Debugging (Optional)
    print(f"Row {i}: x = {x}, y = {y}")

    # Prediction
    y_pred = model.predict_one(x)
    predictions.append(y_pred)

    # Training
    model.learn_one(x, y)

# Save the model
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)

# Add predictions to the DataFrame and save to CSV
consumi_elettrici_2023['Predicted_Consumo'] = predictions
consumi_elettrici_2023.to_csv('consumi_elettrici_2023_predictions.csv', index=False)
print("Predizioni salvate in 'consumi_elettrici_2023_predictions.csv'")
```
